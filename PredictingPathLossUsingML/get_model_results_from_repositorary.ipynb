```python
import pandas as pd
import requests

# URL of the raw JSON file on GitHub
url = "https://raw.githubusercontent.com/daniu54/data-seminar-LoRaWAN/refs/heads/inital-data-analysis/PredictingPathLossUsingML/results/model_stats.json"

# Fetch the raw content of the file
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Read JSON content into a DataFrame
    file_contents = response.text
    model_stats = pd.read_json(file_contents, orient="records", lines=True)
    print(model_stats)
else:
    print(f"Failed to fetch data: {response.status_code}")
```

```python
print(model_stats.columns)

model_stats.sort_values(by="r2", ascending=False, inplace=True)
model_stats.sort_values(by="mse", inplace=True)

print(model_stats[["id", "r2", "mse", "model"]])
```

```python
random_forrest_model = model_stats[model_stats['id'] == 'random_forrest_regressor_default']

assert len(random_forrest_model) > 0, "Should contain that model"

random_forrest_model = random_forrest_model.iloc[0]

print(random_forrest_model[["id", "r2", "mse", "model", "features", "unused_features"]])

used_features = list(random_forrest_model["features"])
unused_features = list(random_forrest_model["unused_features"])

print(used_features)
print(unused_features)
```
